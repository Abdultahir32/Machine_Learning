```python
import numpy as np 
import random 
import time 
import os 
```

```python
n_states = 5 
actions = [-1,1]
q = np.zeros((n_states,len(actions)))
```

```python
alpha = 0.1
gamma = 0.9   #gamma
epsilon = 0.3
episode = 20
```

```python
def show_world(state):
    world = ['üíñ'] * n_states
    world[state] = 'üò∂‚Äçüå´Ô∏è'
    print(" ".join(world))
    print()
```

```python
#training
for ep in range(episode):
    state = 0
    total_reward = 0
    print(f"Episodes {ep+1}")
    time.sleep(0.5)
    while state !=4:
        print(f"Episode {ep+1}")
        show_world(state)
        time.sleep(0.3)
        #exploration, and Explotation
    if random.uniform(0,1) < epsilon:
        action = random.choice(actions)
    else:
        action = actions[np.argmax(Q[state])]
    next_state = max(0, min(4, state+ action))
    reward = 10 if next_state == 4 else -1
    a_index = actions.index(action)
    Q[state, a_index] += alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, a_index])
    state = next_state
    total_reward += reward
    print(f"Epsilon {ep+1}  completed ‚úÖ | Total Reward: {total_reward}\n")
    show_world(state)
    time.sleep(1)
print("\nüëç Training Finished!")
print("Learned Q-Table:")
print(Q)
```
