```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv("HousingData (3).csv")
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df[df['CRIM'].isnull()]
```

```python
for i in df:
    if df[i].isnull().any():
        df[i] = df[i].fillna(df[i].mean())
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
X = df.drop(columns=['MEDV'])
y = df['MEDV']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st = StandardScaler()
x_scaled = st.fit_transform(X)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=7)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred_test = knn.predict(x_test)
y_pred_test
```

```python
y_pred_train = knn.predict(x_train)
y_pred_train
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_pred_train)
```

```python
r2_score(y_test,y_pred_test)
```

```python
score = {}
score_t = {}
for i in range(1,16):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_pred_test = knn.predict(x_test)
    y_pred_train = knn.predict(x_train)
    score[i] = r2_score(y_test,y_pred_test)
    score_t[i] = r2_score(y_train,y_pred_train)
print(score)
print(score_t)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(score.keys(),score.values(),marker = 'o',label='test')
plt.plot(score_t.keys(),score_t.values(),marker='o',color = 'k',label='train')
plt.legend();
```
