# Scraping Data for page 1

```python
Name = []
Price = []
Description = []
url_page = "https://www.flipkart.com/search?q=mobile%20under%2050000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
r = requests.get(url_page)
soup = BeautifulSoup(r.text,"lxml")
# print(soup)

#Scrapping Names

names = soup.find_all("div",class_="KzDlHZ")
# print(names)

for i in names:
    Name.append(i.text)

#Scrapping price 

price = soup.find_all("div",class_="Nx9bqj_4b5DiR")
# print(names)

for i in price:
    # print(i.text)
    Price.append(i.text)

#scrapping the Description

description = soup.find_all("ul",class_="G4BRas")
# print(names)

for i in description:
    # print(i.text)
    Description.append(i.text)



print(Name)
# print(Price)



```
