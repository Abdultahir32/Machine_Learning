```python
from sklearn.datasets import make_regression
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples = 4,n_features=1,n_informative=1,n_targets = 1,noise=80,random_state=13)
```

```python
plt.scatter(X,y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

# Gradient Decent

```python
# lets assuming value of m = 78.35
# lets assuming value of b = 0
y_pred = ((78.35 * X) +0).reshape(4)
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a', label= 'b =  0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0
loss_slope = -2*np.sum(y - m*X.ravel()-b)
loss_slope
```

```python
# lets take learning rate 0.1
lr=0.1
step_size = loss_slope*lr
step_size
```

```python
# calculating the new intercept 
b = b-step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label = 'b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# iteration
loss_slope = -2 *np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
y_pred2 = ((78.35 * X) +b).reshape(4) 
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',linestyle = ':',label = 'b = {}'.format(b))
plt.plot(X,y_pred1,color = '#A3E4D7',label='b=40.9277634')
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```
