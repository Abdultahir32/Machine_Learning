```python
from sklearn.datasets import make_regression
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
x, y = make_regression(n_samples = 4,n_features = 1,n_informative = 1,n_targets =1, noise = 80, random_state = 13)
```

```python
plt.scatter(x,y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red')
```

```python
y_pred = ((78.35* x) +0).reshape(4)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = 'red',label = 'OLS')
plt.plot(x,y_pred,color='green',label='b =0')
plt.legend()
plt.show()
```

```python
b=0
loss_slope = -2*np.sum(y-x.ravel()-b)
loss_slope
```

```python
lr =0.1
step_size = loss_slope *lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred = ((78.35*x)+b).reshape(4)
y_pred
```

```python
b=0
m=78.35
lr=0.1
epochs = 100
for i in range(epochs):
    loss_slope = -2 * np.sqrtr(y -m*x.ravel()-b)
    b = b-(lr*loss_slope)
    y_pred = m*x+b
    print(b)
    plt.plot(x,y_pred)
    plt.plot()
    
```

```python
b=0
m=78.35
lr=0.01
epochs = 100
for i in range(epochs):
    loss_slope = -2 * np.sum(y-m*x.ravel()-b)
    b = b-(lr*loss_slope)
    y_pred = m*x+b
    print(b)
    plt.plot(x,y_pred)
    plt.plot(x,reg.predict(x),color='k',label = 'OLS')
```
