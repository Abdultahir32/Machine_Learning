```python
import nlpcloud    #nlpcloud se Api leni hai
```

```python
class NlpApp:
    def __init__(self):
        self.database = {}
        self.first_menu()

    def first_menu(self):
        first_input=input("press 1 to Register\nPress 2 to Login\nPress 3 Anything to exit")

        if first_input=='1':
            self.register()
        elif first_input=='2':
            self.login()
        else:
            exit()

    def second_menu(self):
        second_input = input('''
        1. NER
        2. Language Detection
        3. Sentiment Analysis
        4. Logout
        
        ''')
        
        if second_input == '1':
            print('You are using NER')
            self.ner()
        elif second_input=='2':
            print('You are using Language Detection')
        elif second_input=='3':
            print('You are using Sentiment Analysis')

        else:
            exit()


    
    def register(self):
        name = input('enter your name')
        email = input('enter your email')
        password = input('enter your password')

        if email in self.database:
            print('Email Already Exist')

        else:
            self.database[email] = [name,password]
            print('Registered Sucessfully')
            print(self.database)
            self.login()
            
    def login(self):
        email = input('enter your email')
        password = input('enter your password')

        if email in self.database:
            if self.database[email][1]==password:
                print('Login sucessfull')
                self.second_menu()
            else:
                print('Wrong Password')
                self.login()
        else:
            print('Email not Found')
            self.first_menu()

    def ner(self):
        import nlpcloud
        client = nlpcloud.Client("gpt-oss-120b", "f209db12cb195e1ed334096dd9d3d9b8a5294fa4", gpu=True)
        t = input('Enter Entity')
        s = input('Enter text')
        response = client.entities(s,searched_entity=t)
        print(response)
        self.second_menu()

                
```

```python
ob = NlpApp()
```
